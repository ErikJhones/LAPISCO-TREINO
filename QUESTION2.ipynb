2) Faça um merge de dois dataframes e crie uma nova coluna para indicar a que dataframe pertence
cada linha.

```python
#bibliotecas usadas
import pandas as pd
import numpy as np
import math
```

```python
#lendo bases de dados
df1 = pd.read_csv("titanic.csv")
df2 = pd.read_csv("iris.csv")
```

```python
df1.head()
```

```python
df2.head()
```

```python
#copiando 10 dados de cada dataset para df1 sub e df2 sub
df1_sub = df1.head(10)
df2_sub = df2.head(10)

#craindo uma lista com os dois dataframes
frames = [df1_sub, df2_sub]

#concatenando os dois dataframes
result = pd.concat(frames)

# Criando nova coluna
result['DATASET_ORIGINAL'] = 'NaN'
result['idex'] = range(20) #coluna com novos indices

#definindo a coluna idex como index
result.set_index('idex', inplace=True)
```

```python
#se linha tiver id não número, então é titanic. se não é iris
for index, row in result.iterrows():
    if math.isnan(row['Id']) == True:
        result.loc[index,'DATASET_ORIGINAL'] = 'titanic' 
    else:
        result.loc[index,'DATASET_ORIGINAL'] = 'iris' 
        
```
